# Carga de datos a traves de la funci√≥n OPEN

```python
# abrir fichero para lectura
data3 = open (minpath + "/" + "customer-churn-model/Customer Churn Model.txt", "r")
type(data3)
data3
```

```python
# cada vez que se ejecuta lee la siguiente linea
# strip limpia espacios
# split indica el separador de campos
cols = data3.readline().strip().split(",")
# saca el numero de campos
n_cols = len(cols)
n_cols
# muestra el registro
cols
```

```python
#crear un diccionario que almacene la clave (contador que se incrementa de 0 a n)
#como valor asociado a esa clave los valores del registro leido. 

# contador de filas
counter=0
#Inicializa el diccionario
main_dict={}
# Por cada registro
for col in cols:
    # inicializa el diccionario poniendo como clave cada campo de la cabecera
    main_dict[col]=[]
main_dict
```

```python
# por cada linea de datos leida
for line in data3:
    # en values guarda los datos sin espacios e indicando que estan separados por coma
    values=line.strip().split(",")
    # desde i=0 hasta el total de columnas
    for i in range(len(cols)):
        # en la clave [x] del diccionario guarda los valores de la columna [x] del dataset
        main_dict[cols[i]].append(values[i])
    # aumenta el contador de filas    
    counter += 1

print ("El dataset tiene %d filas y %d columnas" %(counter, n_cols))
```

```python
# sacar claves del diccionario
main_dict.keys()
```

```python
# sacar un valor por la clave
print (main_dict["Account Length"])
```

```python
# almaceno en un dataframe el contenido del diccionario
df3  = pd.DataFrame(main_dict)
df3.head()
```

## Lectura y escritura de ficheros 

```python
#Ficheros de lectura y escritura
infile = minpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = minpath + "/" + "customer-churn-model/Tab Customer Churn Model.txt" 
```

```python
# abre los ficheros indicando el modo ("r" o "w") y los asigna a variables
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        # por cada linea del ficnhero de entrada
        for line in infile1:
            # limpia la linea
            fields = line.strip().split(",")
            # Une los campos por tabulador
            outfile1.write("\t".join(fields))
            # pone salto de linea
            outfile1.write("\n")
```

```python
# leemos el fichero de salida indicando que esta separado por tabulador
df4 = pd.read_csv(outfile, sep = "\t")
df4.head()
```

# Ejercicio de carga de datos con urllib3

```python
# uso del paquete urllib3 para leer datos de url
import urllib3
medals_url = "http://winterolympicsmedals.com/medals.csv"
http = urllib3.PoolManager()
r = http.request("GET", medals_url)
r.status

# devuelve el valor tipo bytes y lo paso a string
response = r.data.decode(encoding="utf-8")

# separo las lineas por el separador \n
lines = response.strip().split('\n')

# La primera linea es la cabecera
cols = lines[0].strip().split(',')

#diccionario que almacene la clave (contador que se incrementa de 0 a n)
#como valor asociado a esa clave los valores del registro leido. 
# contador de filas
counter=0
#Inicializa el diccionario
medals_dict={}
# Por cada registro
for col in cols:
    # inicializa el diccionario poniendo como clave cada campo de la cabecera
    medals_dict[col]=[]

medals_dict

# por cada linea de datos leida
for line in lines[1:]:
    # en values guarda los datos sin espacios e indicando que estan separados por coma
    values=line.strip().split(",")
    # desde i=0 hasta el total de columnas
    for i in range(len(cols)):
        # en la clave [x] del diccionario guarda los valores de la columna [x] del dataset
        medals_dict[cols[i]].append(values[i])
    # aumenta el contador de filas    
    counter += 1

#print ("El dataset tiene %d filas y %d columnas" %(counter, n_cols))
df3  = pd.DataFrame(medals_dict)
df3.head()
```
